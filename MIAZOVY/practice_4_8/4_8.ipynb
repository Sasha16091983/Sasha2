```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression, make_classification
```

```python
# Генерація даних для регресії
n_samples_reg = 1000 * 10  # кількість даних для регресії (1000 * номер журналу)
X_reg, y_reg = make_regression(n_samples=n_samples_reg, n_features=1, noise=0.1, random_state=42)

# Перевіримо перші кілька значень
print(X_reg[:5], y_reg[:5])
```

```python
# Генерація даних для класифікації
n_samples_clf = 1000 * 10  # кількість даних для класифікації (1000 * номер журналу)
n_classes_clf = 2  # кількість класів (менше або рівне 4)

# Вказуємо 2 інформативні ознаки, 0 надлишкових і 0 повторюваних
X_clf, y_clf = make_classification(
    n_samples=n_samples_clf, 
    n_features=2, 
    n_informative=2,  # Дві інформативні ознаки
    n_redundant=0,    # Жодних надлишкових ознак
    n_repeated=0,     # Жодних повторюваних ознак
    n_classes=n_classes_clf,
    n_clusters_per_class=2,  # Кількість кластерів на клас
    random_state=42
)

# Перевіримо перші кілька значень
print(X_clf[:5], y_clf[:5])



```

```python
# Візуалізація даних для регресії
plt.figure(figsize=(8, 6))
plt.scatter(X_reg, y_reg, alpha=0.6)
plt.title('Generated Data for Regression')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.show()
```

```python
# Візуалізація даних для класифікації
plt.figure(figsize=(8, 6))
plt.scatter(X_clf[:, 0], X_clf[:, 1], c=y_clf, cmap='viridis', alpha=0.6)
plt.title('Generated Data for Classification')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Class')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers, models
from sklearn.metrics import mean_squared_error

# 1. Генерація штучних даних
np.random.seed(42)
X_reg = np.random.rand(1000, 1) * 10           # Фіча від 0 до 10
y_reg = 2.5 * X_reg + np.random.randn(1000, 1) # Ціль з шумом

# 2. Розділення на навчальні та тестові дані
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# 3. Побудова моделі
model_reg = models.Sequential([
    layers.Dense(64, activation='relu', input_dim=1),
    layers.Dense(32, activation='relu'),
    layers.Dense(1)  # Вихід для регресії
])

# 4. Компіляція моделі
model_reg.compile(optimizer='adam', loss='mse')

# 5. Навчання моделі
history = model_reg.fit(X_train_reg, y_train_reg, epochs=50, batch_size=32, verbose=1, validation_split=0.1)

# 6. Оцінка на тестових даних
y_pred = model_reg.predict(X_test_reg)
mse = mean_squared_error(y_test_reg, y_pred)
print("Середньоквадратична помилка (MSE):", mse)

# 7. Побудова графіку
plt.scatter(X_test_reg, y_test_reg, label="Справжні значення", alpha=0.6)
plt.scatter(X_test_reg, y_pred, label="Прогноз", alpha=0.6)
plt.legend()
plt.xlabel("X")
plt.ylabel("y")
plt.title("Прогноз нейронної мережі")
plt.show()

```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers, models

# 1. Створення штучних даних для класифікації
np.random.seed(42)
X_clf = np.random.rand(1000, 2) * 10  # 1000 зразків, 2 ознаки
y_clf = (X_clf[:, 0] + X_clf[:, 1] > 10).astype(int)  # Класифікація: 1 якщо сума ознак більше 10, інакше 0

# 2. Розділення на навчальні та тестові дані
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_clf, y_clf, test_size=0.2, random_state=42)

# 3. Створення моделі для задачі класифікації
model_clf = models.Sequential([
    layers.Dense(64, activation='relu', input_dim=2),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')  # Вихід для класифікації (2 класи)
])

# 4. Компіляція моделі
model_clf.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 5. Навчання моделі
history_clf = model_clf.fit(X_train_clf, y_train_clf, epochs=50, batch_size=32, validation_data=(X_test_clf, y_test_clf))

# 6. Оцінка моделі
loss_clf, accuracy_clf = model_clf.evaluate(X_test_clf, y_test_clf)
print(f"Класифікаційна модель - Втрата на тестових даних: {loss_clf}, Точність: {accuracy_clf}")

```

```python
# Візуалізація історії навчання класифікації
plt.plot(history_clf.history['accuracy'], label='Train Accuracy')
plt.plot(history_clf.history['val_accuracy'], label='Validation Accuracy')
plt.title('Точність на тренуванні та тестуванні по епохах для класифікації')
plt.xlabel('Епохи')
plt.ylabel('Точність')
plt.legend()
plt.show()

plt.plot(history_clf.history['loss'], label='Train Loss')
plt.plot(history_clf.history['val_loss'], label='Validation Loss')
plt.title('Втрати на тренуванні та тестуванні по епохах для класифікації')
plt.xlabel('Епохи')
plt.ylabel('Втрати')
plt.legend()
plt.show()

```

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression, make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, accuracy_score, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Завдання 1: Генерація даних

# Для регресії
n_samples_reg = 1000 * 10  # кількість даних для регресії (1000 * номер журналу)
X_reg, y_reg = make_regression(n_samples=n_samples_reg, n_features=2, noise=0.1, random_state=42)

# Для класифікації
n_samples_clf = 1000 * 10  # кількість даних для класифікації (1000 * номер журналу)
n_classes_clf = 10 + 2  # кількість класів (номер журналу + 2)
X_clf, y_clf = make_classification(n_samples=n_samples_clf, n_features=2, n_classes=n_classes_clf, random_state=42)

# Візуалізація згенерованих даних

    Dense(64, input_dim=2, activation='relu'),

    Dense(1)

model_reg.compile(optimizer='adam', loss='mse')

history_reg = model_reg.fit(X_train_reg, y_train_reg, epochs=50, batch_size=32, validation_split=0.2, verbose=0)

model_clf = Sequential([

model_clf.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Навчання моделі для класифікації
history_clf = model_clf.fit(X_train_clf, y_train_clf, epochs=50, batch_size=32, validation_split=0.2, verbose=0)

# Завдання 3: Оцінка моделей

# Оцінка моделі регресії
y_pred_reg = model_reg.predict(X_test_reg)
mse = mean_squared_error(y_test_reg, y_pred_reg)

y_pred_clf = np.argmax(model_clf.predict(X_test_clf), axis=1)
accuracy = accuracy_score(y_test_clf, y_pred_clf)
