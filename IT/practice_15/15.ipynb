```python
import json

def linear_search_in_json(json_file, key, value):
    """
    Лінійний пошук у файлі JSON за заданим ключем і значенням.
    Повертає список знайдених об'єктів або порожній список, якщо немає збігів.
    """
    with open(json_file, 'r', encoding='utf-8') as f:
        data = json.load(f)

    # Припускаємо, що потрібні об'єкти знаходяться в полі "users"
    users = data.get("users", [])
    results = []
    for user in users:
        if user.get(key) == value:
            results.append(user)
    return results

# Перевірка роботи функції:
found_users = linear_search_in_json('data.json', 'role', 'user')
print("Знайдені користувачі з role='user':", found_users)
```

```python
import json

def build_dict_from_json(json_file, key='id'):
    """
    Зчитує дані з JSON, повертає словник, де ключ - це обране поле (наприклад, 'id'),
    а значення - відповідний об'єкт.
    """
    with open(json_file, 'r', encoding='utf-8') as f:
        data = json.load(f)

    users = data.get("users", [])
    data_dict = {user[key]: user for user in users if key in user}
    return data_dict

# Приклад використання:
users_dict = build_dict_from_json('data.json', key='id')
print("Швидкий пошук користувача з ID=2:", users_dict.get(2))
```

```python
import xml.etree.ElementTree as ET

def search_user_in_xml(xml_file, key, value):
    """
    Пошук користувача в XML файлі за заданим атрибутом і значенням.
    Повертає список знайдених користувачів або порожній список, якщо немає збігів.
    """
    tree = ET.parse(xml_file)
    root = tree.getroot()

    # Знаходимо всіх користувачів у елементі <users>
    users = root.find('users').findall('user')
    
    results = []
    for user in users:
        # Пошук за атрибутами (наприклад, 'id' або 'role') або текстовим значенням
        if user.get(key) == value:
            results.append(user.text)  # Додаємо ім'я користувача до результатів

    return results

# Приклад використання:
found_users = search_user_in_xml('data.xml', 'role', 'user')
print("Знайдені користувачі з role='user':", found_users)
```

```python
import xml.etree.ElementTree as ET

def search_in_xml(xml_file, role_value):
    """
    Пошук усіх елементів <user> з атрибутом role=role_value.
    Повертає список словників із даними про знайдених користувачів.
    """
    tree = ET.parse(xml_file)  # Парсимо XML файл
    root = tree.getroot()  # Отримуємо корінь дерева

    result = []  # Список для збереження результатів пошуку
    users_element = root.find('users')  # Знаходимо елемент <users>
    
    if users_element is not None:
        for user_element in users_element.findall('user'):  # Перебираємо кожен <user>
            if user_element.get('role') == role_value:  # Порівнюємо значення атрибута role
                # Формуємо словник з атрибутів та текстового значення <user>
                user_info = {
                    "id": user_element.get("id"),  # Атрибут id
                    "name": user_element.text,  # Текстовий вміст елемента <user> (ім'я користувача)
                    "role": user_element.get("role")  # Атрибут role
                }
                result.append(user_info)  # Додаємо знайденого користувача до результату
    return result

# Перевірка функції
found_xml_users = search_in_xml('data.xml', 'user')
print("Користувачі із role='user':", found_xml_users)
```

```python
def binary_search_in_list(sorted_list, target_id):
    """
    Бінарний пошук у відсортованому списку за id.
    Повертає користувача з заданим id або None, якщо не знайдено.
    """
    left, right = 0, len(sorted_list) - 1
    
    while left <= right:
        mid = (left + right) // 2  # Знаходимо середину
        if sorted_list[mid]['id'] == target_id:
            return sorted_list[mid]  # Знайдений елемент
        elif sorted_list[mid]['id'] < target_id:
            left = mid + 1  # Шукаємо в правій частині
        else:
            right = mid - 1  # Шукаємо в лівій частині
    
    return None  # Якщо елемент не знайдений

# Приклад відсортованого списку користувачів
users_sorted = [
    {"id": 1, "name": "Alice", "role": "admin"},
    {"id": 2, "name": "Bob", "role": "user"},
    {"id": 3, "name": "Charlie", "role": "user"},
    {"id": 4, "name": "Diana", "role": "manager"}
]

# Перевірка функції
print("Знайдений користувач:", binary_search_in_list(users_sorted, 3))
```

```python
import time
import json
import xml.etree.ElementTree as ET
import random

# Генерація тестових даних
def generate_json_data(n):
    data = {
        "users": [{"id": i, "name": f"User {i}", "role": "user"} for i in range(1, n+1)]
    }
    with open('data.json', 'w', encoding='utf-8') as f:
        json.dump(data, f)

def generate_xml_data(n):
    root = ET.Element("root")
    users = ET.SubElement(root, "users")
    for i in range(1, n+1):
        user = ET.SubElement(users, "user", id=str(i), role="user")
        user.text = f"User {i}"
    tree = ET.ElementTree(root)
    tree.write("data.xml", encoding="utf-8")

# Лінійний пошук в JSON
def linear_search_in_json(json_file, target_id):
    with open(json_file, 'r', encoding='utf-8') as f:
        data = json.load(f)
    users = data.get("users", [])
    for user in users:
        if user['id'] == target_id:
            return user
    return None

# Бінарний пошук в JSON
def binary_search_in_list(sorted_list, target_id):
    left, right = 0, len(sorted_list) - 1
    while left <= right:
        mid = (left + right) // 2
        if sorted_list[mid]['id'] == target_id:
            return sorted_list[mid]
        elif sorted_list[mid]['id'] < target_id:
            left = mid + 1
        else:
            right = mid - 1
    return None

# Пошук за словником (dict) в JSON
def dict_search_in_json(json_file, target_id):
    with open(json_file, 'r', encoding='utf-8') as f:
        data = json.load(f)
    users_dict = {user['id']: user for user in data['users']}
    return users_dict.get(target_id)

# Лінійний пошук в XML
def linear_search_in_xml(xml_file, target_id):
    tree = ET.parse(xml_file)
    root = tree.getroot()
    users_element = root.find('users')
    for user_element in users_element.findall('user'):
        if user_element.get('id') == str(target_id):
            return {
                "id": user_element.get("id"),
                "name": user_element.text,
                "role": user_element.get("role")
            }
    return None

# Бінарний пошук в XML
def binary_search_in_xml(xml_file, target_id):
    tree = ET.parse(xml_file)
    root = tree.getroot()
    users_element = root.find('users')
    users = [{"id": int(user_element.get('id')), "name": user_element.text, "role": user_element.get('role')} 
             for user_element in users_element.findall('user')]
    users.sort(key=lambda x: x['id'])  # Сортуємо список для бінарного пошуку
    return binary_search_in_list(users, target_id)

# Пошук за словником (dict) в XML
def dict_search_in_xml(xml_file, target_id):
    tree = ET.parse(xml_file)
    root = tree.getroot()
    users_element = root.find('users')
    users_dict = {int(user_element.get('id')): {
                    "id": user_element.get("id"),
                    "name": user_element.text,
                    "role": user_element.get("role")} 
                  for user_element in users_element.findall('user')}
    return users_dict.get(target_id)

# Генерація даних
n = 10000  # 10,000 елементів
generate_json_data(n)
generate_xml_data(n)

# Тестування часу на різних методах пошуку

# Лінійний пошук в JSON
start_time = time.time()
linear_search_in_json('data.json', 5000)
end_time = time.time()
print(f"Час лінійного пошуку в JSON: {end_time - start_time} секунд")

# Бінарний пошук в JSON
start_time = time.time()
binary_search_in_list(sorted([user for user in json.load(open('data.json'))['users']], key=lambda x: x['id']), 5000)
end_time = time.time()
print(f"Час бінарного пошуку в JSON: {end_time - start_time} секунд")

# Пошук за словником в JSON
start_time = time.time()
dict_search_in_json('data.json', 5000)
end_time = time.time()
print(f"Час пошуку за словником в JSON: {end_time - start_time} секунд")

# Лінійний пошук в XML
start_time = time.time()
linear_search_in_xml('data.xml', 5000)
end_time = time.time()
print(f"Час лінійного пошуку в XML: {end_time - start_time} секунд")

# Бінарний пошук в XML
start_time = time.time()
binary_search_in_xml('data.xml', 5000)
end_time = time.time()
print(f"Час бінарного пошуку в XML: {end_time - start_time} секунд")

# Пошук за словником в XML
start_time = time.time()
dict_search_in_xml('data.xml', 5000)
end_time = time.time()
print(f"Час пошуку за словником в XML: {end_time - start_time} секунд")
```

```python
import json
import re

def search_users_by_multiple_criteria(json_file, role, name_prefix):
    with open(json_file, 'r', encoding='utf-8') as f:
        data = json.load(f)
    users = data.get("users", [])
    
    result = []
    for user in users:
        if user['role'] == role and re.match(f"^{name_prefix}", user['name']):
            result.append(user)
    return result

# Перевірка
found_users = search_users_by_multiple_criteria('data.json', 'user', 'B')
print("Знайдені користувачі з role='user' та name, що починається з 'B':", found_users)
```

```python
import xml.etree.ElementTree as ET

def index_xml_data(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()
    users_element = root.find('users')
    
    # Створюємо індекс: словник за id
    users_dict = {}
    for user_element in users_element.findall('user'):
        user_id = int(user_element.get('id'))
        users_dict[user_id] = {
            "id": user_element.get("id"),
            "name": user_element.text,
            "role": user_element.get("role")
        }
    
    return users_dict

# Приклад пошуку після створення індексу
users_dict = index_xml_data('data.xml')
print("Знайдений користувач з id=5000:", users_dict.get(5000))
```

```python
import csv

def generate_csv_data(n):
    with open('data.csv', mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(["id", "name", "role"])
        for i in range(1, n+1):
            writer.writerow([i, f"User {i}", "user"])

generate_csv_data(10000)  # Створюємо CSV з 10000 користувачів
```

```python
import csv
import re

def search_users_in_csv(csv_file, role, name_prefix):
    result = []
    with open(csv_file, mode='r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        for row in reader:
            if row['role'] == role and re.match(f"^{name_prefix}", row['name']):
                result.append(row)
    return result

# Перевірка
found_users_csv = search_users_in_csv('data.csv', 'user', 'B')
print("Знайдені користувачі в CSV з role='user' та name, що починається з 'B':", found_users_csv)
```
